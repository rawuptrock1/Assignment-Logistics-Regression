```python
#Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.
'''Answer:
The SLR is  supervised learning algorithm used to model the relationship between a single input feature (X) and a continuous target variable (y).
The Purpose: The idea of SLR is to explain how changes done in the independent variable influence the dependent variable and not via versa. Alos to use this relationship for prediction and inference.
Mathematically speking, SLR  fits the best possible straight line passing through the data points such that the difference between the actual values and predicted values is minimized.
	Prediction
SLR is widely used to predict future or unknown values of a dependent variable based on known values of the independent variable.
	Understanding Relationships
It helps in understanding wether an independent variable has a positive, negative, or no impact on the dependent variable.
	Quantifying Impact
Slope of the reg line describe how much the dependent variable changes for a unit change in the independent variable.
	Trend Analysis
SLR is useful for identifying trends in data, such as growth, decline, or stability over time.
	Decision Making
Businesses and researchers use SLR to make data-driven decisions, such as budgeting, forecasting demand, and performance evaluation.
In summary, Simple Linear Regression is a foundational technique that combines simplicity, interpretability, and predictive power, making it essential in statistics and machine learning.
'''
```

```python

#Question 2: What are the key assumptions of Simple Linear Regression?
'''Answer: SLRmodel to produce reliable and unbiased results, certain assumptions that must be satisfied. These assumptions ensure the validity of predictions, confidence intervals, and hypothesis tests. Based on standard regression theory studied in callssroom, for an SLR model to be reliable, the following must hold true:
	Linearity: The relationship between X and Y must be representable by a straight line.
	Independence of error: The observations must be independent of each other.
	Homoscedasticity: The variance of the error terms  must be constant across all levels of the independent variable.
	Normality of Residuals: The errors should follow a normal distribution, which is imp for hypothesis testing and calculating confidence intervals.
	No Extreme Outliers
Large outliers can disproportionately influence the regression line and distort predictions.
Violation of these assumptions does not always invalidate the model, but it can reduce accuracy and interpretability, especially for statistical inference.
'''
```

```python

#Question 3: Write the mathematical equation for a simple linear regression model and explain each term.
'''Answer: The mathematical equation for the SLR model is:
y = b0 + b1*x + e
Explanation of Each Term
	y
The dependent (response) variable that the model aims to predict.
	x
The independent (predictor) variable used to explain changes in y.
	b0 (Intercept)
The value of ywhen x=0.
It represents the baseline level of the dependent variable.
	b1 (Slope or Coefficient)
Indicates how much y changes for a one-unit increase in x.
A positive slope indicates a direct relationship, while a negative slope indicates an inverse relationship.
	e(Error Term)
Represents random noise or unexplained variation not captured by the model.
This equation separates systematic variation (explained by x) from random variation (error term).
'''
```

```python
#Question 5: What is the method of least squares in linear regression?
'''Answer: The Ordinary Least Squares (OLS) method is the optimization strategy used to calculate the best values for b0 and b1 It works by minimizing the Sum of Squared Errors (SSE), which is the sum of the squares of the vertical distances between the actual data points and the regression line. Squaring the errors ensures that large outliers are penalized more than small deviations.
Core Idea
OLS determines the values of β_0and β_1such that the sum of squared residuals is minimized.
SSE=∑(y_i-y ̂_i )^2 
Where:
	y_i= actual value
	y ̂_i= predicted value
Why Squared Errors?
	Ensures positive values
	Penalizes large errors more heavily
	Makes optimization mathematically tractable
Importance
	Produces a unique best-fit line
	Forms the foundation of many machine learning algorithms
	Guarantees minimum prediction error under regression assumptions
OLS ensures the regression line is as close as possible to all data points collectively.

'''
```

```python
#Question 6: What is Logistic Regression? How does it differ from Linear Regression?
'''Answer: Logistic Regression is used for Classification tasks where the outcome is categorical (e.g., 0 or 1).
How Logistic Regression Works
Instead of predicting raw values, it predicts the probability that an observation belongs to a particular class using the sigmoid function.
σ(z)=1/(1+e^(-z) )
Key Differences:
	Outcome: Linear Regression predicts a continuous value; Logistic Regression predicts the probability of an observation belonging to a specific class.
	Activation Function: Logistic Regression uses the Sigmoid (Logistic) Function to map any real-valued number into a range between 0 and 1.
	Decision Boundary: Logistic Regression applies a threshold (usually 0.5) to the probability to decide the final class.
Understand Differences from Linear Regression in tabular format 
Aspect	Linear Regression	Logistic Regression
Output	Continuous	Probability (0–1)
Task	Prediction	Classification
Function	Straight line	Sigmoid curve
Loss	MSE	Log loss
Decision	Direct value	Threshold-based
Logistic Regression is widely used in spam detection, disease diagnosis, and fraud detection.

'''
```

```python
#Question 7: Name and briefly describe three common evaluation metrics for regression models.
'''Answer: To evaluate how well the model fits the data, we use:
1. Mean Absolute Error (MAE)
	Average of absolute differences between actual and predicted values.
	Easy to interpret.
	Less sensitive to outliers.
2. Mean Squared Error (MSE)
	Average of squared errors.
	Penalizes larger errors more.
	Useful for optimization algorithms.
3. Root Mean Squared Error (RMSE)
	Square root of MSE.
	Expressed in the same units as the target variable.
	Commonly used for performance reporting.
Each metric provides a different perspective on model accuracy.
'''
```

```python
#Question 8: What is the purpose of the R-squared metric in regression analysis?
'''Answer: The R-squared (R^2) metric, also known as the Coefficient of Determination, measures the "Goodness of Fit". It represents the proportion of the variance in the dependent variable that is explained by the independent variable(s).
R^2=1-(SS_res)/(SS_tot )

Purpose
	Indicates goodness of fit
	Shows explanatory power of the model
	Helps compare different models
Interpretation
	R^2=0.85: 85% variance explained
	R^2=0: No explanatory power
However, a high R^2does not guarantee causation or correctness.
'''
```

```python
#Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
'''Answer: The following code reflects the implementation structure found in your resources using sklearn.linear_model:
Cde:
'''
import numpy as np
from sklearn.linear_model import LinearRegression


X = np.array([[1], [2], [3], [4], [5]]) 
y = np.array([45000, 50000, 60000, 72000, 81000])


regressor = LinearRegression()

regressor.fit(X, y)

slope = regressor.coef_[0]
intercept = regressor.intercept_
print(f"Slope (Beta 1): {slope}")
print(f"Intercept (Beta 0): {intercept}")
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, regressor.predict(X), color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression: Salary vs Experience')
plt.legend()
plt.show()

```
